```python
import numpy as np
import os
import random
from numba import jit
from numpy import arange
import copy
import time
```

```python
#T = 16
#number of jumps 
class chess:
    """
    Class to describe status keys are 0-15 of the chess position
    each keys associate with a summation dict, keys are summations, values are probabilities
    """
    def __init__(self):
        self.status = {}
        self.next = {}
            #keys are 0-15 postions
        for i in range(16):
            self.next[i] = {}   
            self.status[i] = {}
            #where keys are sumamtions and values are probabilities
        self.status[0][0] = 1.0
            #initial
        self.final_prob = {}
            #for final probability use
        self.possible_moves = {}
            #possible moves for all positions 
        self.generate_moves()
            #generate dict for possible_moves
         
    def generate_moves(self):
        """generate possible move number and positions after for each positions"""
        for i in range(16):
            num_temp = 0
            possible_list = []
            self.possible_moves[i] = {}
            for j in arange(max(-2,0-i%4),min(3,4-i%4)):
                for k in arange(max(-2,0-int(i/4)),min(3,4-int(i/4))):
                    if (abs(j)+abs(k) == 3):
                        num_temp += 1
                        possible_list.append(i+j+4*k)
            self.possible_moves[i]["num"] = num_temp
            self.possible_moves[i]["list"] = possible_list
    #tested
         
    def reset(self):
        """reset status = all blank"""
        self.status = {}
        for i in range(16):
            self.status[i] = {}
    def set_status_next(self):
        """set status to next status"""
        self.status = copy.deepcopy(self.next)
    def reset_next(self):
        """reset next status = all blank"""
        self.next = {}
        for i in range(16):
            self.next[i] = {}      
  
    def iteration(self): 
        """method to iterate to next steps, generate next dict, then copy to status"""
        self.reset_next()
        for i in range(16):
            #possible moves associate with positions
            n = float(self.possible_moves[i]["num"])
            list_pos = self.possible_moves[i]["list"]
            for key in self.status[i].keys():
                for new_pos in list_pos:
                    #assign new probability
                    new_key = key+new_pos
                    if new_key in self.next[new_pos].keys():
                        self.next[new_pos][new_key] += self.status[i][key]/n
                    else:
                        self.next[new_pos][new_key] = self.status[i][key]/n
        self.set_status_next()
     
    def get_final_prob(self):
        """sum over all positions to obtain dict of summation as key and prob as values"""
        for i in range(16):
            for key in self.status[i].keys():
                if key in self.final_prob.keys():
                    self.final_prob[key] += self.status[i][key]
                else:
                    self.final_prob[key] = self.status[i][key]
    
    
                  
                    
                
```
